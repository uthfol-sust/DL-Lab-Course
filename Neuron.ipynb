```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
def Sigmoid(x):
    return 1/(1+np.exp(-x))

def Sigmoid_derivative(x):
    s = Sigmoid(x)
    return s*(1-s)
```

```python
class Perceptron:

    def __init__(self, learning_rate=0.01, n_iter=100):
        self.lr = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y):

        X = np.asarray(X, dtype=np.float64)
        y = np.asarray(y)

        n_samples, n_features = X.shape

        self.weights = np.zeros(n_features)
        self.bias = 0.0

        for epoch in range(self.n_iter):

            for idx, x_i in enumerate(X):

                z = np.dot(x_i, self.weights) + self.bias

                y_pred = Sigmoid(z)

                error = y_pred - y[idx]

                gradient = error * Sigmoid_derivative(z)

                self.weights -= self.lr * gradient * x_i
                self.bias -= self.lr * gradient

            if (epoch + 1) % 10 == 0:
                print(f"Epoch {epoch+1}/{self.n_iter} "f"Loss={np.mean((self.predict_proba(X)-y)**2):.4f}" )

    def predict_proba(self, X):

        X = np.asarray(X, dtype=np.float64)

        z = np.dot(X, self.weights) + self.bias

        return Sigmoid(z)

    def predict(self, X):

        probs = self.predict_proba(X)

        return (probs >= 0.5).astype(int)
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")
df.head()
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
df.head()
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df.head()
```

```python
df.fillna(df.mean(), inplace=True)
df.head()
```

```python
x = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
import sklearn.model_selection as model_selection

x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.2)
```

```python


p = Perceptron(
    learning_rate=0.01,
    n_iter=100
)

p.fit(x_train, y_train)
```

```python
preds=p.predict(x_test)
```

```python
acc = accuracy_score(y_test, preds)
acc
```
